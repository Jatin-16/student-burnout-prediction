```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_squared_error,r2_score
```

```python
df = pd.read_csv("student-mat.csv")
df["sex"] = (df["sex"].str.lower() == "m").astype(int)
df
```

```python
clean_df = df[["studytime","traveltime","failures","absences","health","freetime","goout","G1","G2","G3"]]
clean_df
```

```python
x_train,x_test,y_train,y_test = train_test_split(clean_df.drop(["G3"],axis=1),clean_df["G3"],test_size=0.4,random_state=42)
```

```python
scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_train_scaled
```

```python
model = LinearRegression()
model.fit(x_train_scaled,y_train)
```

```python
y_train_pred = model.predict(x_train_scaled)
```

```python
mse = mean_squared_error(y_train,y_train_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_train,y_train_pred)
print(f"mse: {mse},  rmse: {rmse},  r2: {r2}")
```

```python
y_test_pred = model.predict(scaler.transform(x_test))
```

```python
mse = mean_squared_error(y_test,y_test_pred)
rmse = np.sqrt(mse)
r2 = r2_score(y_test,y_test_pred)
print(f"mse: {mse},  rmse: {rmse},  r2: {r2}")
```

```python
import pickle

# Save trained model
with open("model//burnout_model.pkl", "wb") as f:
    pickle.dump(model, f)

# Save fitted scaler
with open("model//scaler.pkl", "wb") as f:
    pickle.dump(scaler, f)

# Save feature order (very important)
features = ["studytime","traveltime","failures","absences",
            "health","freetime","goout","G1","G2"]

with open("model//features.pkl", "wb") as f:
    pickle.dump(features, f)

print("Model, scaler, and features saved successfully.")
```
